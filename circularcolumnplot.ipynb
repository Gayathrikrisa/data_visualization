```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('sqldata/sslc_score.csv')
df.head()
```

```python
fig=plt.figure()
ax=fig.add_subplot()
ax.bar(df['Subject'],df['Score'])
ax.set_xlabel('Subjects')
ax.set_ylabel('Score')
ax.set_title('Score Card')
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(polar=True)
ax.bar(df['Subject'],df['Score'])
ax.set_xlabel('Subjects')
ax.set_ylabel('Score')
ax.set_title('Score Card')
plt.show()
```
