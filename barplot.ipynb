```python
df=pd.read_csv('sqldata/students_score.csv')
df
```

```python
tamil=df['Score'].loc[df['Subject']=='Tamil']
english=df['Score'].loc[df['Subject']=='English']
maths=df['Score'].loc[df['Subject']=='Maths']
science=df['Score'].loc[df['Subject']=='Science']
social=df['Score'].loc[df['Subject']=='Social']
print(tamil)

subjects=df['Subject'].unique()
students=df['Student'].unique()

eng_btm=np.array(tamil)
mat_btm=eng_btm+np.array(english)
sci_btm=mat_btm+np.array(maths)
soc_btm=sci_btm+np.array(science)
print(eng_btm)

fig=plt.figure(dpi=120)
ax=fig.add_subplot(1,1,1)
ax.bar(students,tamil,bottom=tamil,color='r')
ax.bar(students,english,bottom=eng_btm,color='g')
ax.bar(students,maths,bottom=mat_btm,color='b')
ax.bar(students,science,bottom=sci_btm,color='g')
ax.bar(students,social,bottom=soc_btm,color='y')


for index,value in enumerate(tamil.to_list()):
    ax.text(index,value,value)
for index,value in enumerate(zip(eng_btm,english.to_list())):
    ax.text(index,sum(value),sum(value))
for index,value in enumerate(zip(mat_btm,maths.to_list())):
    ax.text(index,sum(value),sum(value))
for index,value in enumerate(zip(sci_btm,science.to_list())):
    ax.text(index,sum(value),sum(value))
for index,value in enumerate(zip(soc_btm,social.to_list())):
    ax.text(index,sum(value),sum(value))

ax.set_xticks(students)
ax.set_xticklabels(students)

ax.set_xlabel('Subjects',fontdict=xfont,labelpad=10)
ax.set_ylabel('Score',fontdict=yfont,labelpad=0)
ax.set_title('Score-Card',fontdict=title)
ax.legend(subjects,loc=(1,1))
plt.show()


```
