```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
xfont={'family':'serif',
       'color':'r',
       'weight':'bold',
       'size':14}
yfont={'family':'serif',
       'color':'r',
       'weight':'bold',
       'size':14}
highlight={'family':'serif',
           'color':'darkviolet',
           'weight':'bold',
           'size':14
           }
```

```python
df=pd.read_csv('sqldata/sslc_score.csv')
df.head()
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.barh(df['Subject'],df['Score'])

ax.set_xlabel("Score",fontdict=xfont,labelpad=0)
ax.set_ylabel("Subject",fontdict=yfont,labelpad=0,loc='center')
ax.set_title("score card")

df.sort_values(by=['Score'],ascending=False,inplace=True)
plt.show()
```
