```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('sqldata/covid_death.csv')
df.head()
```

```python

x=df['Country']
y=df['Death']

fig=plt.figure()
ax=fig.add_subplot()
ax.plot(x,y,label='values',color='b')
ax.set_xlabel('Country')
ax.set_ylabel('Death Rate')
ax.set_title('Covid Death Status')
ax.set_xticklabels(x,rotation=90)
plt.show()
```

```python
x=df['Country'][:6]
y=df['Death'][:6]

fig=plt.figure()
ax=fig.add_subplot(1,1,1,polar=True)
ax.plot(x,y,label='values',color='b')
# ax.set_xlabel('Country')
# ax.set_ylabel('Death Rate')
ax.set_title('Covid Death Status')
ax.set_xticklabels(x,rotation=90)
plt.show()
```
