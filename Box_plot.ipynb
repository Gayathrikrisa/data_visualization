```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('sqldata/cars.csv')
fig=plt.figure()
ax=fig.add_subplot()
ax.boxplot(df['Weight'])
ax.set_xticklabels(['weight'],rotation=0)
ax.set_ylabel("values")
ax.set_title("Outlier Analysis Box plot")
```

```python
df=pd.read_csv('sqldata/Iris.csv')
df.head()

```

```python
fig=plt.figure()
ax=fig.add_subplot()
ax.scatter(df['PetalLengthCm'],df['PetalWidthCm'])
ax.set_xlabel('Peteal Length',labelpad=0)
ax.set_ylabel('Petal Width',labelpad=0, loc='center')
ax.set_title('Iris Flowers')
plt.show()
```

```python
species=df['Species'].unique()
species
```

```python
df['colors']=df['Species'].map({'Iris-setosa':'r','Iris-versicolor':'g','Iris-virginica':'b'})
```

```python
fig=plt.figure()
ax=fig.add_subplot()
size=df['SepalWidthCm']*80
ax.scatter(data=df,x='PetalLengthCm',y='PetalWidthCm',s=size,c='colors',alpha=0.5)
ax.set_xlabel('Peteal Length',labelpad=0)
ax.set_ylabel('Petal Width',labelpad=0, loc='center')
ax.set_title('Iris Flowers')
plt.show()
```
