```python
xfont={'color':'blue',
       'weight':'bold',
       'size':14}
yfont={'color':'red',
       'weight':'bold',
       'size':16}
title={'color':'green',
       'weight':'bold',
       'size':14}
x=nb.arange(30)
sinx=nb.sin(x)

fig=plt.figure()
ax=fig.add_subplot()
ax.plot(x.cumsum(),label='Vals',color='blue')
ax.plot(sinx.cumsum(),label='sin vals',color='green')

ax.set_xlabel('X Axis',fontdict=xfont,labelpad=0)
ax.set_ylabel('Y Axis',fontdict=yfont,labelpad=0,loc='center')
ax.set_title('Lion Plot',fontdict=title)

ax.legend(bbox_to_anchor=(0.8,-0.2),ncol=2)
plt.show()

```

```python
df=pd.read_csv('sqldata/covid_death.csv')
df.head()
```

```python
x=df['Country']
y=df['Death']

xfont={'color':'blue',
       'weight':'bold',
       'size':14}
yfont={'color':'red',
       'weight':'bold',
       'size':16}
title={'color':'green',
       'weight':'bold',
       'size':14}

fig=plt.figure(dpi=150)
ax=fig.add_subplot(1,1,1)
ax.plot(x,y,label='Vals',color='blue',marker='o',linestyle='dotted')
ax.set_xticklabels(x,rotation=90)

ax.set_xlabel('Country',fontdict=xfont)
ax.set_ylabel('DeathRate',fontdict=yfont)
ax.set_title('Covid Death',fontdict=title)
plt.show()
```

```python
df=pd.read_csv('sqldata/covid_tamilnadu.csv')
df.head()
```

```python
district=df['District']
positive=df['Positive']
discharged=df['Discharged']
active=df['Active']
death=df['Death']

fig=plt.figure(figsize=(16,5))
ax=fig.add_subplot(1,1,1)

ax.plot(district,positive,label='Positive',color='red')
ax.plot(district,discharged,label='Discharged',color='green')
# ax.plot(district,active,label='Active',color='k')
# ax.plot(district,death,label='Death',color='red')


ax.set_xticklabels(district,rotation=90)
ax.set_xlabel('District',fontdict=xfont)
ax.set_ylabel('Count',fontdict=yfont)
ax.set_title('Covid-Tamilnadu',fontdict=title)

ax.legend(loc='best')
plt.show()
```
