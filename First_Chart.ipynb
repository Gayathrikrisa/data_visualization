```python
import numpy as np
import matplotlib.pyplot as plt
x=np.random.randn(500)
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(x.cumsum())
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(x.cumsum())
ticks=ax.set_xticks([0,100,200,300,400])
labels=ax.set_xticklabels(['zero','quater','pre-mid','post-mid','Third'])
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
ax.plot(x.cumsum())

xticks=ax.set_xticks([0,100,200,300,400])
yticks=ax.set_yticks([-10,0,10,20])


xlabels=ax.set_xticklabels(['Zero','Quater','Pre-mid','Post-mid','Third'])
ylabels=ax.set_yticklabels(['Negative','Zero','Positive','High'])
```

```python
fig=plt.figure(dpi=120)# dot per inch
ax=fig.add_subplot(1,1,1)
ax.plot(x.cumsum(),linestyle="solid",color='r')

xticks=ax.set_xticks([0,100,200,300,400])
tyicks=ax.set_yticks([-10,0,10,20])
ax.set_title("Report for days variation",loc='Right',fontdict={'color':'m','fontweight':'bold'})

xlabels=ax.set_xticklabels(['Zero','Quater','Premid','Pst-mid','Third'],fontdict={'fontsize':14,'fontweight':800,'color':'b'},rotation=45)

ylabels=ax.set_yticklabels(['Negative','Zero','Positive','High'])
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.random.randn(500)
sinx=np.sin(x)

fig=plt.figure()
ax=fig.add_subplot()
ax.plot(x.cumsum(),label='x values')
ax.plot(sinx.cumsum(),label='Sinx values')

xfont={'color':'r','size':10}
yfont={'color':'g','size':10,'rotation':90}
title={'color':'b', 'size':18,'style':'italic'}
# legend_prop={'weight':'semibond','size':12, 'style':'underline'}

ax.set_xlabel('Count',fontdict=xfont,labelpad=0)
ax.set_ylabel('Values',fontdict=yfont,labelpad=0)
ax.set_title('Random Plot',fontdict=title)

ax.legend(loc='upper right')
```

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

ax.plot(x.cumsum(),label='X values',c='b')
ax.plot(sinx.cumsum(),label='Sin Values',c='#22adcf')

xfont={'color':'r','weight':'bold','size':12}
yfont={'color':'g','weight':'bold','size':12}
title={'color':'m','weight':'semibold','size':15}

ax.set_xlabel('Count',fontdict=xfont,labelpad=10)
ax.set_ylabel('Values',fontdict=yfont,labelpad=10)
ax.set_title('Random Plot Values',fontdict=title)

legend=ax.legend(bbox_to_anchor=(1,1),ncol=2)
text=legend.get_texts()
# for values in text:
#     values.set_color("#3dcd68")
text[0].set_color('blue')
text[1].set_color('green')
```

```python
#Annotation
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

ax.plot(x.cumsum(),label='X values',c='b')
ax.plot(sinx.cumsum(),label='Sin Values',c='#22adcf')

xfont={'color':'r','weight':'bold','size':12}
yfont={'color':'g','weight':'bold','size':12}
title={'color':'m','weight':'semibold','size':15}

ax.set_xlabel('Count',fontdict=xfont,labelpad=10)
ax.set_ylabel('Values',fontdict=yfont,labelpad=10)
ax.set_title('Random Plot Values',fontdict=title)

legend=ax.legend(bbox_to_anchor=(1,1),ncol=2)
text=legend.get_texts()
# for values in text:
#     values.set_color("#3dcd68")
text[0].set_color('blue')
text[1].set_color('green')
peak1=ax.annotate('peak1',(120,5),rotation=-0)
peak2=ax.annotate('peak2',(250,9),rotation=0)
peak3=ax.annotate('peak3',(430,8),rotation=0)
rect=plt.Rectangle((120,0),width=20,height=2,color='b')
ax.add_patch(rect)
rect=plt.Rectangle((240,8),width=20,height=2,color='r')
ax.add_patch(rect)
rect=plt.Rectangle((420,6),width=40,height=2, color='g')
ax.add_patch(rect)
```
