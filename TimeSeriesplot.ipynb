```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
```

```python
df=pd.read_csv('sqldata/bajajfins.csv')
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.head()
```

```python
fig=plt.figure()
ax=fig.add_subplot()
ax.plot(df['Date'],df['High'],label='values',color='green')
ax.set_xticklabels(df['Date'],rotation=90)
ax.set_xlabel('TimeLine')
ax.set_ylabel('Capital gain-high')
ax.set_title('Bajaj fin service')
plt.show()
```
